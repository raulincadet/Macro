```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from datetime import datetime
```

```python
df=pd.read_csv("COFER_07-01-2023 02-51-41-63_timeSeries.csv")
df.head(3)
```

```python
# select only quarterly data
cols=['Country Name','Indicator Name']
[cols.append(i) for i in list(df.columns) if len(i)==6 and i[4]=='Q' ]
cols
df=df[cols]
df.head(3)
```

```python
# select aggregate data for the World, where the indicator is the allocated reserves, claims in different currencies.
ind=[df.index[i] for i in range(df.shape[0]) if (df.loc[i,'Indicator Name'].startswith('Allocated Reserves, Claims')) and (df.loc[i,'Country Name']=='World') ]
df1=df.iloc[ind,:]
df1=df1.reset_index()
df1.head(3)
```

```python
df1['Currency']=[i[30:len(i)-10] for i in df1['Indicator Name']]
df1.head(3)
```

```python
df1=df1.set_index('Currency')
df1=df1.iloc[:,3:]
df1
```

```python
import numpy as np
def all_nan(x):
    import numpy as np
    y=list()
    for i in range(len(x)):
        if np.isnan(x[i])==True:
            y.append(1)
    if len(y)==len(x):
        return True
    else:
        return False
all_nan(list(df1.iloc[0,:]))       
```

```python
df2=df1[[all_nan(df1.iloc[i,:])==False for i in range(df1.shape[0])]]
df2=df2.transpose()
```

```python
# consider ratios instead
import numpy as np
tot_by_period=[np.sum(df2.iloc[i,:]) for i in range(df2.shape[0])] 
```

```python
dd1=df2
# fig,ax=plt.subplots()
dd1['Total']=tot_by_period
dd1.index=pd.to_datetime(list(dd1.index))
dou=pd.DataFrame()
for currency in dd1.columns:
    dou[currency+'_ratio']=dd1[currency]/dd1['Total']
    dou[currency+'_rol']=dou[currency+'_ratio'].rolling(4).mean()
    fig=plt.subplot()
    plt.plot(dou[currency+'_ratio'],color='blue')
    plt.plot(dou[currency+'_rol'],color='orange')
    plt.title(currency)
    plt.show()
```
