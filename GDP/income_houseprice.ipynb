```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
dhpi1=pd.read_excel("HPI_PO_state.xls")
dhpi1.head(4)
```

```python
dhpi1[dhpi1.state=="MA"][dhpi1.yr==2023]
```

```python
dhpi=dhpi1[(dhpi1.yr==2023) &(dhpi1.qtr==3)]
dhpi=dhpi.reset_index()
dhpi
```

```python
df_inc=pd.read_csv('SQINC4__ALL_AREAS_1948_2023 (2).csv')

# Select rows related to Personal income
df_inc1=df_inc[df_inc.Description=='Personal income (millions of dollars, seasonally adjusted) ']
df_inc1=df_inc1.reset_index()
df_inc1.head(3)
```

```python
dinc=pd.DataFrame({'States':df_inc1['GeoName'],'Income':df_inc1['2023:Q3']*1000000})
dinc=dinc.set_index('States')
dinc.head()
```

```python
# correct states names where an * appear next to their names.
States=list()
for i in dinc.index:
    if '*' in i:
        states.append(i[:-1])
    else:
        States.append(i)
    
dinc.index=[i.strip() for i in States] # remove space in some states name
dinc.shape,dinc.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing the table
url = 'https://www.bu.edu/brand/guidelines/editorial-style/us-state-abbreviations/'

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table 
table = soup.find('table')

# Convert the HTML table to a DataFrame
df_states = pd.read_html(str(table))[0]

df_states.head(4)
```

```python
df_states.columns=['States','Traditional abbreviation','state']
df_states=df_states.loc[1:,:]
df_states=df_states.set_index('States')
df_states=df_states[['state']]
df_states.head(4)
```

```python
d3=pd.merge(dinc,df_states,left_index=True,right_index=True)
d3.shape
```

```python
df_pop=pd.read_excel('NST-EST2023-POP.xlsx',header=3,nrows=57) # Source: https://www.census.gov/data/tables/time-series/demo/popest/2020s-state-total.html
df_pop.columns=['States','Estimates base','Population 2020','Population 2021','Population 2022','Population 2023']
df_pop=df_pop.set_index('States') 
df_pop.head()
```

```python
# to remove the dot before some states names
z=list()
for i in df_pop.index:
    if '.' in i:
        z.append(i[1:])
    else:
        z.append(i)
    
df_pop.index=z

# df_pop=df_pop.reset_index()
# df_pop=df_pop.rename(columns={'index': 'States'})
# df_pop=df_pop.set_index('States')
df_pop.head(6)
```

```python
d4=pd.merge(d3,df_pop,left_index=True,right_index=True)
d4['Personnal income 2023Q3 (% population 2023)']=d4['Income']/d4['Population 2023']
d4.head(4)
```

```python
# Relate region to state data
df_geo=pd.read_excel('state-geocodes-v2014_modifyCRL.xlsx',sheet_name='Modify_Raulin')
df_geo=df_geo.set_index('States')
d5=pd.merge(d4,df_geo,left_index=True,right_index=True)
d5['region']=[i[:-7] for i in d5['Region']]
```

```python
d6=pd.merge(d5,dhpi,on='state')
d6.head()
```

### Graphic

```python
region_colors = {
    'Northeast': 'darkviolet',
    'Midwest': 'darkturquoise',    
    'South': 'green',      
    'West': 'saddlebrown',       
}

# regions=list(set(d5.region))  # list of regions
# colors = [region_colors[region] for region in regions]
```

```python


fig=plt.subplot()
regionplotted=list()
for i in range(d6.shape[0]):
    if d6.loc[i,'region'] not in regionplotted:
        plt.scatter(x=d6.loc[i,'Personnal income 2023Q3 (% population 2023)'],y=d6.loc[i,'index_nsa'],color=region_colors[d6.loc[i,'region']],label=d6.loc[i,'region'],alpha=0.65,s=90)
        regionplotted.append(d6.loc[i,'region'])
    else:
        plt.scatter(x=d6.loc[i,'Personnal income 2023Q3 (% population 2023)'],y=d6.loc[i,'index_nsa'],color=region_colors[d6.loc[i,'region']],alpha=0.65,s=90)
    
    plt.text(x=d6.loc[i,'Personnal income 2023Q3 (% population 2023)']+750,y=d6.loc[i,'index_nsa'],s=d6.loc[i,'state'])
plt.axvline(x=np.mean(d6['Personnal income 2023Q3 (% population 2023)']),color='red',linestyle='--')
plt.axhline(y=np.mean(d6['index_nsa']),color='red',linestyle='--')
plt.xlabel('Personal income in USD')
plt.ylabel('House price index')
plt.title('House Price index (HPI) & Per Capita Personal Income\n by US States (2023 Quarter 3)',pad=55,size=14)
plt.gca().spines[['top','left','right','bottom']].set_visible(False)
plt.text(x=np.mean(d6['Personnal income 2023Q3 (% population 2023)'])*0.98,y=np.mean(d6['index_nsa'])*0.62,color='red',s='Mean',rotation=90)
plt.text(x=np.min(d6['Personnal income 2023Q3 (% population 2023)']),y=np.mean(d6['index_nsa'])*1.01,color='red',s='Mean HPI')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=len(region_colors))
plt.text(x=np.min(d6['Personnal income 2023Q3 (% population 2023)']),y=np.min(d6['index_nsa'])-165,
         s="Source: By Raulin L. Cadet, with personal income data from the U.S.\n Bureau of Economic Analysis, House Price Index data from the Federal\n Housing Finance Agency. Population estimations, from the U.S. Census\n Bureau, is used to calculate per capita personal income.")

```
