Compare government and private expenditures per capita, by country

```python
# import Python modules
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# import data
# https://www.who.int/data/gho/data/indicators/indicator-details/GHO/domestic-private-health-expenditure-(pvt-d)-as-percentage-of-current-health-expenditure-(che)-(-)
df=pd.read_excel("data_Health_expenditures.xlsx",usecols='A,B,E,H,J,AD',skiprows=range(0,2)) 
df.head(5)
```

```python
# select only the last year available data: 2021
df2=df[df['Period']==2021]
df2=df2.sort_values('Value')
df2.head(10)
```

```python
fig = plt.figure()  # Create a figure object
ax = fig.add_subplot()  # Create a subplot

# Plot horizontal bar chart
bars = ax.barh(df2.tail(10)['Location'], df2.tail(10)['Value'])

# Add value labels next to the bars
for bar in bars:
    width = bar.get_width()  # Get the width of the bar
    ax.text(width, bar.get_y() + bar.get_height() / 2,  # Position the text
            f'{width:.2f}%',  # Format the text to show the value with two decimal places
            va='center', ha='left', fontsize=10)  # Vertical alignment center, horizontal alignment left

ax.spines[['top', 'bottom', 'left', 'right']].set_visible(False)  # Remove borders
ax.set_xticks([])  # Remove x-axis ticks
plt.title("Top Countries by Domestic Private Health Expenditure \n(% of Total Health Expenditure) in 2021")
plt.text(x=-5, y=-1, s='Source: Raulin Cadet, with data from the World Health Organization.', fontsize=10)

plt.show()  
```
